# Feature Analyze

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Data

```python
total = pd.read_csv("data/total.csv")
```

```python
total.head()
```

## Viasual every Feature according label 

```python
colors = np.where(total["label"]==1,'r','-')
colors[total["label"]==0] = 'g'
```

```python
for col in total.columns.values[:-1]:
    total.plot.scatter(col, "label", title = col, c=colors, yticks=[0,1,1])
```
